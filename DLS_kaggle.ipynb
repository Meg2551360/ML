## Применение линейных моделей (3 балла)

1) Обработайте данные для того, чтобы к ним можно было применить LogisticRegression. Т.е. отнормируйте числовые признаки, а категориальные закодируйте с помощью one-hot-encoding'а.

(более сложный вариант) Если вы будете использовать кроссвалидацию, то преобразования данных и LogisticRegression нужно соединить в один Pipeline с помощью make_pipeline, как это делалось во втором семинаре. Для one-hot-encoding'a можно испльзовать комбинацию LabelEncoder + OneHotEncoder (сначала превращаем строчки в числа, а потом числа првращаем в one-hot вектора.)

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, classification_report
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.pipeline import Pipeline
```

```python
X = data.drop(['Churn'], axis=1)
y = data['Churn']
```

```python
# Обработка данных: нормировка числовых признаков, категориальные закодированы с помощью one-hot-encoding
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), cat_cols),
        ('num', StandardScaler(), make_column_selector(dtype_include=['int64', 'float64']))
    ])
```

```python
X_norm = preprocessor.fit_transform(X)
```

```python
X_train, X_val, y_train, y_val= train_test_split(X, y.values, test_size=0.20,shuffle=True, random_state=37)
```

```python
# LogisticRegressionCV
pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegressionCV(
        Cs=[100, 10, 1, 0.1, 0.01, 0.001],  # Значения для гиперпараметра C
        cv=5,  # Количество фолдов для кросс-валидации
        scoring='roc_auc',  # Метрика для оценки
        class_weight='balanced',  # Балансировка классов
        random_state=42,
        refit=True  # Переобучение модели на всем датасете с лучшими параметрами
    ))
])

pipe.fit(X_train, y_train)

y_pred = pipe.predict(X_val)
y_pred_proba = pipe.predict_proba(X_val)[:, 1]

print(classification_report(y_val, y_pred))
print(f"ROC-AUC: {roc_auc_score(y_val, y_pred_proba):.4f}")
```

```python
# Создание Pipeline
pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=42))
])

# Определение сетки параметров для GridSearchCV
param_grid = [
    {
        'classifier': [LogisticRegression(solver='liblinear', class_weight='balanced', random_state=47)],
        'classifier__C': [100, 10, 1, 0.1, 0.01, 0.001],  # Значения для гиперпараметра C
        'classifier__penalty': ['l1', 'l2']  # Тип регуляризации
    },
    {
        'classifier': [RandomForestClassifier(class_weight='balanced', random_state=47)],
        'classifier__max_depth': [2, 5, 10, 20],
        'classifier__n_estimators': list(range(1, 21, 2)),
        'classifier__min_samples_leaf': list(range(1, 10, 3)),
        'classifier__min_samples_split': [2, 5, 10]  # Исправлено: min_samples_split >= 2
    },
    {
        'classifier': [DecisionTreeClassifier(class_weight='balanced', random_state=47)],
        'classifier__max_depth': [2, 5, 10, 20],
        'classifier__min_samples_leaf': list(range(1, 10, 3)),
        'classifier__min_samples_split': [2, 5, 10]  # Исправлено: min_samples_split >= 2
    }
]

# Создание GridSearchCV
grid_search = GridSearchCV(
    pipe,
    param_grid,
    scoring='roc_auc',  # Метрика для оценки
    cv=10,  # Количество фолдов для кросс-валидации
    refit=True  # Переобучение модели на всем датасете с лучшими параметрами
)

# Обучение GridSearchCV на обучающей выборке
grid_search.fit(X_train, y_train)

# Лучшие параметры
print(f"Лучшие параметры: {grid_search.best_params_}")

# Предсказание на тестовой выборке
y_pred = grid_search.predict(X_val)
y_pred_proba = grid_search.predict_proba(X_val)[:, 1]  # Вероятности для ROC-AUC

# Оценка модели
print(classification_report(y_val, y_pred))
print(f"ROC-AUC: {roc_auc_score(y_val, y_pred_proba):.4f}")
```

```python
fpr, tpr, thresholds = roc_curve(y_val, grid_search.predict_proba(X_val)[:, 1])

plt.figure(figsize=(6, 6))
plt.step(fpr, tpr, where='post')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('Кривая ROC')
plt.show()
```

Наилучшее качество ROC-AUC = 0.8511 было найдено с помощью GridSearchCV.  
Best model: LogisticRegression (class_weight='balanced', random_state=47, solver='liblinear'), 'classifier__C': 1,'classifier__penalty': 'l2', количество эпох обучения - 10.

## Применение градиентного бустинга (2 балла)

Если вы хотите получить баллы за точный ответ, то стоит попробовать градиентный бустинг. Часто градиентный бустинг с параметрами по умолчанию даст вам 80% результата за 0% усилий.

Мы будем использовать catboost, поэтому нам не надо кодировать категориальные признаки. Catboost сделает это сам (в .fit() надо передать cat_features=cat_cols). А численные признаки нормировать для моделей, основанных на деревьях не нужно.

1) Разделите выборку на train/valid. Протестируйте catboost cо стандартными параметрами.

2) Протестируйте разные занчения параметров количества деревьев и learning_rate'а и выберите лучшую по метрике ROC-AUC комбинацию.

(Дополнительно) Есть некоторые сложности с тем, чтобы использовать CatBoostClassifier вместе с GridSearchCV, поэтому мы не просим использовать кроссвалидацию. Но можете попробовать)

```python
%pip install catboost
```

```python
from catboost import CatBoostClassifier
from collections import Counter
```

```python
# разделение выше, перезапускала сеанс
model_Cat = CatBoostClassifier(
    cat_features=cat_cols,  # Указываем категориальные признаки
    random_state=42,
    verbose=0  # Отключаем вывод логов
)

# Обучение модели
model_Cat.fit(X_train, y_train)

# Предсказание на валидационной выборке
y_pred_proba = model_Cat.predict_proba(X_val)[:, 1]

# Оценка модели по метрике ROC-AUC
roc_auc = roc_auc_score(y_val, y_pred_proba)
print(f"ROC-AUC (CatBoost по умолчанию): {roc_auc:.4f}")
```

```python
model_Cat_gr = CatBoostClassifier(
    random_state=42,
    verbose=0
)

# учитываем дисбаланс классов
class_counts = Counter(y_train)
majority_class = max(class_counts, key=class_counts.get)
minority_class = min(class_counts, key=class_counts.get)

scale_pos_weight = class_counts[majority_class] / class_counts[minority_class]
print(f"scale_pos_weight: {scale_pos_weight:.2f}")

# Добавляем в сетку гиперпараметров
param_grid = {
    'depth': [4, 6, 8],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'n_estimators': [100, 200, 500],
    'scale_pos_weight': [scale_pos_weight]  # Учитываем дисбаланс
}

# Создаем GridSearchCV
grid_search = GridSearchCV(
    estimator=model_Cat_gr,
    param_grid=param_grid,
    scoring='roc_auc',
    cv=3,
    n_jobs=-1
)

# Запускаем поиск с передачей cat_features в fit
grid_search.fit(X_train, y_train, cat_features=cat_cols)

# Выводим лучший результат
print("Лучшие параметры:", grid_search.best_params_)
print("Лучший ROC-AUC:", grid_search.best_score_)
```

**Вывод:**
