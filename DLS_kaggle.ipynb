```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, classification_report
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.pipeline import Pipeline
```

```python
X = data.drop(['Churn'], axis=1)
y = data['Churn']
```

```python
#  :   ,     one-hot-encoding
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), cat_cols),
        ('num', StandardScaler(), make_column_selector(dtype_include=['int64', 'float64']))
    ])
```

```python
X_norm = preprocessor.fit_transform(X)
```

```python
X_train, X_val, y_train, y_val= train_test_split(X, y.values, test_size=0.20,shuffle=True, random_state=37)
```

```python
# LogisticRegressionCV
pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegressionCV(
        Cs=[100, 10, 1, 0.1, 0.01, 0.001],  #    C
        cv=5,  #    -
        scoring='roc_auc',  #   
        class_weight='balanced',  #  
        random_state=42,
        refit=True  #        
    ))
])

pipe.fit(X_train, y_train)

y_pred = pipe.predict(X_val)
y_pred_proba = pipe.predict_proba(X_val)[:, 1]

print(classification_report(y_val, y_pred))
print(f"ROC-AUC: {roc_auc_score(y_val, y_pred_proba):.4f}")
```

```python
#  Pipeline
pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=42))
])

#     GridSearchCV
param_grid = [
    {
        'classifier': [LogisticRegression(solver='liblinear', class_weight='balanced', random_state=47)],
        'classifier__C': [100, 10, 1, 0.1, 0.01, 0.001],  #    C
        'classifier__penalty': ['l1', 'l2']  #  
    },
    {
        'classifier': [RandomForestClassifier(class_weight='balanced', random_state=47)],
        'classifier__max_depth': [2, 5, 10, 20],
        'classifier__n_estimators': list(range(1, 21, 2)),
        'classifier__min_samples_leaf': list(range(1, 10, 3)),
        'classifier__min_samples_split': [2, 5, 10]  # : min_samples_split >= 2
    },
    {
        'classifier': [DecisionTreeClassifier(class_weight='balanced', random_state=47)],
        'classifier__max_depth': [2, 5, 10, 20],
        'classifier__min_samples_leaf': list(range(1, 10, 3)),
        'classifier__min_samples_split': [2, 5, 10]  # : min_samples_split >= 2
    }
]

#  GridSearchCV
grid_search = GridSearchCV(
    pipe,
    param_grid,
    scoring='roc_auc',  #   
    cv=10,  #    -
    refit=True  #        
)

#  GridSearchCV   
grid_search.fit(X_train, y_train)

#  
print(f" : {grid_search.best_params_}")

#    
y_pred = grid_search.predict(X_val)
y_pred_proba = grid_search.predict_proba(X_val)[:, 1]  #   ROC-AUC

#  
print(classification_report(y_val, y_pred))
print(f"ROC-AUC: {roc_auc_score(y_val, y_pred_proba):.4f}")
```

```python
fpr, tpr, thresholds = roc_curve(y_val, grid_search.predict_proba(X_val)[:, 1])

plt.figure(figsize=(6, 6))
plt.step(fpr, tpr, where='post')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title(' ROC')
plt.show()
```

```python
from catboost import CatBoostClassifier
from collections import Counter
```

```python
#  ,  
model_Cat = CatBoostClassifier(
    cat_features=cat_cols,  #   
    random_state=42,
    verbose=0  #   
)

#  
model_Cat.fit(X_train, y_train)

#    
y_pred_proba = model_Cat.predict_proba(X_val)[:, 1]

#     ROC-AUC
roc_auc = roc_auc_score(y_val, y_pred_proba)
print(f"ROC-AUC (CatBoost  ): {roc_auc:.4f}")
```

```python
model_Cat_gr = CatBoostClassifier(
    random_state=42,
    verbose=0
)

#   
class_counts = Counter(y_train)
majority_class = max(class_counts, key=class_counts.get)
minority_class = min(class_counts, key=class_counts.get)

scale_pos_weight = class_counts[majority_class] / class_counts[minority_class]
print(f"scale_pos_weight: {scale_pos_weight:.2f}")

#    
param_grid = {
    'depth': [4, 6, 8],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'n_estimators': [100, 200, 500],
    'scale_pos_weight': [scale_pos_weight]  #  
}

#  GridSearchCV
grid_search = GridSearchCV(
    estimator=model_Cat_gr,
    param_grid=param_grid,
    scoring='roc_auc',
    cv=3,
    n_jobs=-1
)

#     cat_features  fit
grid_search.fit(X_train, y_train, cat_features=cat_cols)

#   
print(" :", grid_search.best_params_)
print(" ROC-AUC:", grid_search.best_score_)
```
